```python
import sys
import numpy as np
import json
import argparse
import zipfile
import tempfile
import gzip
import shutil

parser = argparse.ArgumentParser(description='Input')
parser.add_argument('-param_file', action='store', dest='param_file',
                    help='param file file')
parser.add_argument('-input', action='store', dest='input_file_path',
                    help='input file path')
parser.add_argument('-output', action='store',dest='output_file_path',
                    help='output file path')
args, unknown = parser.parse_known_args()

args.param_file='processed_files/text8.param.json'
args.input_file_path='files_to_be_compressed/text8/text8'
args.output_file_path='processed_files/text8.npy'
```

```python
with open(args.input_file_path, 'rb') as fp:
    data = fp.read()

print(len(data))
vals = list(set(data))

char2id_dict = {c: i for (i,c) in enumerate(vals)}
id2char_dict = {i: c for (i,c) in enumerate(vals)}

params = {'char2id_dict':char2id_dict, 'id2char_dict':id2char_dict}
with open(args.param_file, 'w') as f:
    json.dump(params, f, indent=4)
```

```python
subsample_ls=[100000,500000,1000000,2000000]
for subsample in subsample_ls:
    data_sub=data[0:subsample]
    fs=open('files_to_be_compressed/text8/text8_sub'+str(subsample), 'wb')
    fs.write(data_sub)
    fs.close()

    out = [char2id_dict[c] for c in data_sub]
    integer_encoded = np.array(out)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    np.save('processed_files/text8_sub'+str(subsample)+'.npy', integer_encoded)

    with open('files_to_be_compressed/text8/text8_sub'+str(subsample), 'rb') as f_in:
        with gzip.open('files_to_be_compressed/text8/text8_sub'+str(subsample)+'.gz', 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
```
